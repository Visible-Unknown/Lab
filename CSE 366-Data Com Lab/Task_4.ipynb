# Digital-digital conversion techniques

```python
# NRZ
import matplotlib.pyplot as plt

def nrz_encode(bits):
    signal = []
    for bit in bits:
        if bit == 0:
            signal.extend([0, 0])
        else:
            signal.extend([1, 1])
    return signal

# Example bits to encode
bits = [0, 1, 0, 1, 1, 0, 1]

# NRZ encoding
encoded_signal = nrz_encode(bits)

# Adjust the encoded signal to properly represent the last bit
# Append the last bit value once more to ensure it's shown completely
encoded_signal.append(encoded_signal[-1])

# Plot the NRZ encoded signal
plt.figure(figsize=(10, 4))
plt.step(range(len(encoded_signal)), encoded_signal, where='post')
plt.ylim([-2, 2])
plt.title('NRZ Encoding')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(linewidth=1)

# Annotate the bitstream
for i, bit in enumerate(bits):
    plt.text(i * 2 + 0.5, 1.5, str(bit), fontsize=12, ha='center')

plt.show()
```

```python
#NRZ-L

import matplotlib.pyplot as plt

def nrzL_encode(bits):
    signal = []
    for bit in bits:
        if bit == 0:
            signal.extend([1, 1])
        else:
            signal.extend([-1, -1])
    return signal

# Example bits to encode
bits = [0, 1, 0, 1, 1, 0, 1]

# NRZ encoding
encoded_signal = nrzL_encode(bits)

# Adjust the encoded signal to properly represent the last bit
# Append the last bit value once more to ensure it's shown completely
encoded_signal.append(encoded_signal[-1])

# Plot the NRZ-L encoded signal
plt.figure(figsize=(10, 4))
plt.step(range(len(encoded_signal)), encoded_signal, where='post')
plt.ylim([-2, 2])
plt.title('NRZ-L Encoding')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid()

# Annotate the bitstream
for i, bit in enumerate(bits):
    plt.text(i * 2 + 0.5, 1.5, str(bit), fontsize=12, ha='center')

plt.show()
```

```python
# Manchester

import matplotlib.pyplot as plt

def manchester_encode(bits):
    signal = []
    for bit in bits:
        if bit == 0:
            signal.extend([1, -1])
        else:
            signal.extend([-1, 1])
    return signal

# Example bits to encode
bits = [0, 1, 0, 1, 1, 0, 1]

# Manchester encoding
encoded_signal = manchester_encode(bits)

# Adjust the encoded signal to properly represent the last bit
# Append the last bit value once more to ensure it's shown completely
encoded_signal.append(encoded_signal[-1])

# Plot the Manchester encoded signal
plt.figure(figsize=(10, 4))
plt.step(range(len(encoded_signal)), encoded_signal, where='post')
plt.ylim([-2, 2])
plt.title('Manchester Encoding')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid()

# Annotate the bitstream
for i, bit in enumerate(bits):
    plt.text(i * 2 + 0.5, 1.5, str(bit), fontsize=12, ha='center')

plt.show()

```

```python
#NRZ

import matplotlib.pyplot as plt

def differential_manchester_encode(bits):
    signal = []
    current_level = 1  # start with high
    for bit in bits:
        if bit == 0:
            signal.extend([-current_level, current_level])
        else:
            signal.extend([current_level, -current_level])
        current_level = signal[-1]
    return signal

# Example bits to encode
bits = [0, 1, 0, 1, 1, 0, 1]

# Differential Manchester encoding
encoded_signal = differential_manchester_encode(bits)

# Adjust the encoded signal to properly represent the last bit
# Append the last bit value once more to ensure it's shown completely
encoded_signal.append(encoded_signal[-1])

# Plot the Differential Manchester encoded signal
plt.figure(figsize=(10, 4))
plt.step(range(len(encoded_signal)), encoded_signal, where='post')
plt.ylim([-2, 2])
plt.title('Differential Manchester Encoding')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(linewidth = 1)

# Annotate the bitstream
for i, bit in enumerate(bits):
    plt.text(i * 2 + 0.5, 1.5, str(bit), fontsize=12, ha='center')

plt.show()
```
