# Digital-analog conversion techniques

```python
#ASK
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
data_bits = np.array([1, 0, 1, 1, 1, 0])  # Binary data sequence
bit_rate = 1  # bits per second
carrier_freq = 10  # Carrier frequency in Hz
sampling_rate = 1000  # Samples per second
amplitude_high = 1
amplitude_low = 0

# Time array
bit_duration = 1 / bit_rate
t = np.arange(0, len(data_bits) * bit_duration, 1 / sampling_rate)

# Carrier signal
carrier_signal = np.sin(2 * np.pi * carrier_freq * t)

# Modulated signal
modulated_signal = np.zeros_like(t)
for i, bit in enumerate(data_bits):
    start_idx = int(i * bit_duration * sampling_rate)
    end_idx = int((i + 1) * bit_duration * sampling_rate)
    if bit == 1:
        modulated_signal[start_idx:end_idx] = amplitude_high * carrier_signal[start_idx:end_idx]
    else:
        modulated_signal[start_idx:end_idx] = amplitude_low

# Plotting
plt.figure(figsize=(20, 6))

# Plot binary data
plt.subplot(3, 1, 1)
plt.step(np.arange(len(data_bits) + 1), np.append(data_bits, data_bits[-1]), where='post', color='red', linewidth=2)
plt.ylim(-0.5, 1.5)
plt.ylabel('Data bits')
plt.title('Binary Data Sequence')
plt.grid(True)

# Plot carrier signal
plt.subplot(3, 1, 2)
plt.plot(t, carrier_signal, color='black')
plt.ylabel('Amplitude')
plt.title('Carrier Signal')
plt.grid(True)

# Plot modulated signal
plt.subplot(3, 1, 3)
plt.plot(t, modulated_signal, color='blue')
plt.ylabel('Amplitude')
plt.xlabel('Time (s)')
plt.title('ASK Modulated Signal')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# FSK
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
data_bits = np.array([1, 0, 1, 0, 1, 0])  # Binary data sequence
bit_rate = 1  # bits per second
carrier_freq_0 = 5  # Carrier frequency for binary 0 in Hz
carrier_freq_1 = 10  # Carrier frequency for binary 1 in Hz
sampling_rate = 1000  # Samples per second

# Time array
bit_duration = 1 / bit_rate
t = np.arange(0, len(data_bits) * bit_duration, 1 / sampling_rate)

# Modulated signal
modulated_signal = np.zeros_like(t)
for i, bit in enumerate(data_bits):
    start_idx = int(i * bit_duration * sampling_rate)
    end_idx = int((i + 1) * bit_duration * sampling_rate)
    if bit == 1:
        modulated_signal[start_idx:end_idx] = np.sin(2 * np.pi * carrier_freq_1 * t[start_idx:end_idx])
    else:
        modulated_signal[start_idx:end_idx] = np.sin(2 * np.pi * carrier_freq_0 * t[start_idx:end_idx])

# Plotting
plt.figure(figsize=(20, 6))

# Plot binary data
plt.subplot(3, 1, 1)
plt.step(np.arange(len(data_bits)+1), np.append(data_bits, data_bits[-1]), where='post', color='red', linewidth=2)
plt.ylim(-0.5, 1.5)
plt.ylabel('Data bits')
plt.title('Binary Data Sequence')
plt.grid(True)

# Plot carrier signals for reference
plt.subplot(3, 1, 2)
# plt.plot(t, np.sin(2 * np.pi * carrier_freq_0 * t), color='black', label='Carrier Signal')
plt.plot(t, np.sin(2 * np.pi * carrier_freq_0 * t), color='black')
plt.ylabel('Amplitude')
plt.title('Carrier Signals')
plt.grid(True)

# Plot modulated signal
plt.subplot(3, 1, 3)
plt.plot(t, modulated_signal, color='blue')
plt.ylabel('Amplitude')
plt.xlabel('Time (s)')
plt.title('FSK Modulated Signal')
plt.grid(True)

plt.tight_layout()
plt.show()
```
