## Ploting Simple Signal

```python
# Required modules
import numpy as np
import matplotlib.pyplot as plt

# Time vector
t = np.linspace(0, 1, 1000)  # 1 second duration, 1000 points

# Simple sinusoidal signal
A1 = 2  # Amplitude
f1 = 5  # Frequency in Hz
y1 = A1 * np.sin(2 * np.pi * f1 * t)

# Simple signal plot
plt.plot(t, y1, label="Simple Sinusoidal Signal")
plt.title("Simple Sinusoidal Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.axhline(0,color='black',linewidth=0.8,linestyle='--')
```
