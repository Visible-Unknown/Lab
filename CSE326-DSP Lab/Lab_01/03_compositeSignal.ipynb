## Plotting Composite Signal

## Required modules

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Time vector

```python
# Time vector
t = np.linspace(0, 1, 1000)  # 1 second duration, 1000 points
```

## Define three signals

```python
A1 = 2  # Amplitude of the second signal
f1 = 2  # Frequency of the second signal
y1 = A1 * np.sin(2 * np.pi * f1 * t)

A2 = 1  # Amplitude of the second signal
f2 = 6  # Frequency of the second signal
y2 = A2 * np.sin(2 * np.pi * f2 * t)

A3 = 0.5  # Amplitude of the third signal
f3 = 12  # Frequency of the third signal
y3 = A3 * np.sin(2 * np.pi * f3 * t)
```

## Composite signal

```python
y_composite = y1 + y2 + y3
```

## Ploting

```python
# Composite signal
plt.plot(t, y_composite, label="Composite Signal", color="r")
plt.title("Composite Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.axhline(0,color='black',linewidth=0.8,linestyle='--')
plt.legend()
plt.show()
```
