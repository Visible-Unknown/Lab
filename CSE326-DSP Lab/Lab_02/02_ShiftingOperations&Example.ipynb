```python
import matplotlib.pyplot as plt

# Define time index and input signal
t = list(range(-4, 6))
s = [2, 3, 1, 5, 2, 2, 2, 2, 2, 2]

# 1. Manual Shift Forward (s(n+3)): Shift Left by 3
def shift_forward(signal, shift):
    shifted = [0] * len(signal)  # Initialize with zeros
    for i in range(len(signal) - shift):
        shifted[i] = signal[i + shift]
    return shifted

s_shifted_forward = shift_forward(s, 3)

# 2. Manual Shift Backward (s(n-2)): Shift Right by 2
def shift_backward(signal, shift):
    shifted = [0] * len(signal)  # Initialize with zeros
    for i in range(shift, len(signal)):
        shifted[i] = signal[i - shift]
    return shifted

s_shifted_backward = shift_backward(s, 2)

# 3. Manual Moving Average Filter
def moving_average(signal, window_size):
    avg_filtered = [0] * len(signal)
    for i in range(len(signal)):
        window = signal[max(0, i - window_size // 2): min(len(signal), i + window_size // 2 + 1)]
        avg_filtered[i] = sum(window) / len(window)
    return avg_filtered

avg_filter = moving_average(s, 4)

# 4. Manual Median Filter
def median_filter(signal, window_size):
    median_filtered = [0] * len(signal)
    for i in range(len(signal)):
        window = signal[max(0, i - window_size // 2): min(len(signal), i + window_size // 2 + 1)]
        sorted_window = sorted(window)
        median_filtered[i] = sorted_window[len(sorted_window) // 2]  # Pick median
    return median_filtered

median_filtered = median_filter(s, 4)

# 5. Manual Cumulative Sum
def cumulative_sum(signal):
    cum_sum = [0] * len(signal)
    sum_value = 0
    for i in range(len(signal)):
        sum_value += signal[i]
        cum_sum[i] = sum_value
    return cum_sum

cumulative_sum_result = cumulative_sum(s)

# Plot all results
plt.figure(figsize=(14, 12))

plt.subplot(3, 2, 1)
plt.stem(t, s)
plt.title("Input Signal s(n)")
plt.xlabel("n")
plt.ylabel("Value")
plt.grid()

plt.subplot(3, 2, 2)
plt.stem(t, s_shifted_forward)
plt.title("Shifted Forward Signal s(n+3)")
plt.xlabel("n")
plt.ylabel("Value")
plt.grid()

plt.subplot(3, 2, 3)
plt.stem(t, s_shifted_backward)
plt.title("Shifted Backward Signal s(n-2)")
plt.xlabel("n")
plt.ylabel("Value")
plt.grid()

plt.subplot(3, 2, 4)
plt.stem(t, avg_filter)
plt.title("Moving Average Filter Output")
plt.xlabel("n")
plt.ylabel("Value")
plt.grid()

plt.subplot(3, 2, 5)
plt.stem(t, median_filtered)
plt.title("Median Filter Output")
plt.xlabel("n")
plt.ylabel("Value")
plt.grid()

plt.subplot(3, 2, 6)
plt.stem(t, cumulative_sum_result)
plt.title("Cumulative Sum Output")
plt.xlabel("n")
plt.ylabel("Value")
plt.grid()

plt.tight_layout()
plt.show()
```
