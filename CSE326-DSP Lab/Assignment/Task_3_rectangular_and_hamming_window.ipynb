Problem statement: Design a Rectangular Window and a Hamming Window, and compare their properties and effects on signal processing applications.

Required Library

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import windows 
```

Implementation

```python
def design_windows(N=51):
    """
    Design and plot a rectangular and hamming window.

    Parameters:
        N: Length of the window
    """
    # Generate windows
    rect_window = windows.boxcar(N)      # Rectangular window
    hamming_window = windows.hamming(N)  # Hamming window

    # Plot
    plt.figure(figsize=(10, 4))
    plt.plot(rect_window, label="Rectangular Window", linestyle='--')
    plt.plot(hamming_window, label="Hamming Window")
    plt.title(f"Window Functions (Length = {N})")
    plt.xlabel("Sample")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

    return rect_window, hamming_window

# Example usage
rect, hamm = design_windows(N=51)
```
