Problem Statement: Design a Band-Pass Filter and a Band-Stop Filter, and examine their performance in isolating or eliminating specific frequency bands.

Required library

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import firwin, freqz
```

Implementation

```python
def design_band_filters(fs=1000, band=[200, 400], numtaps=101):
    """
    Design Band-Pass and Band-Stop FIR filters.

    Parameters:
        fs: Sampling frequency (Hz)
        band: List [low_cutoff, high_cutoff] in Hz
        numtaps: Number of filter taps

    Returns:
        bp_filter: Band-pass filter coefficients
        bs_filter: Band-stop filter coefficients
    """
    # Band-pass filter
    bp_filter = firwin(numtaps, cutoff=band, fs=fs, pass_zero=False, window='hamming')

    # Band-stop filter
    bs_filter = firwin(numtaps, cutoff=band, fs=fs, pass_zero=True, window='hamming')

    # Frequency response
    w_bp, h_bp = freqz(bp_filter, worN=8000, fs=fs)
    w_bs, h_bs = freqz(bs_filter, worN=8000, fs=fs)

    # Plot frequency responses
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(w_bp, 20 * np.log10(np.abs(h_bp)))
    plt.title("Band-Pass Filter Frequency Response")
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Magnitude (dB)")
    plt.grid()

    plt.subplot(1, 2, 2)
    plt.plot(w_bs, 20 * np.log10(np.abs(h_bs)))
    plt.title("Band-Stop Filter Frequency Response")
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Magnitude (dB)")
    plt.grid()

    plt.tight_layout()
    plt.show()

    return bp_filter, bs_filter

# Example usage
bp, bs = design_band_filters(fs=1000, band=[200, 400], numtaps=101)
```
