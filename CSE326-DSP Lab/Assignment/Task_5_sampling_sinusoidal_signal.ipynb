Problem statement: Sample a continuous sinusoidal signal (e.g., x(t)=sin(2Ï€ft)) at a discrete sampling frequency fs like 500 Hz.

Required  Library

```python
import numpy as np
import matplotlib.pyplot as plt
```

Implementation

```python
def sample_sinusoid(signal_freq=50, sampling_freq=500, duration=0.1):
    """
    Plot a continuous sinusoidal signal and its sampled version.

    Parameters:
        signal_freq: Frequency of the sinusoid (Hz)
        sampling_freq: Sampling rate (Hz)
        duration: Duration of the signal (seconds)

    Returns:
        t_samples: Sample times
        x_samples: Sampled values
    """
    # Continuous-time sinusoid (high resolution for smooth plot)
    t_cont = np.linspace(0, duration, 1000)
    x_cont = np.sin(2 * np.pi * signal_freq * t_cont)

    # Sampled version
    t_samples = np.arange(0, duration, 1 / sampling_freq)
    x_samples = np.sin(2 * np.pi * signal_freq * t_samples)

    # Plot: Continuous sinusoid
    plt.figure(figsize=(10, 4))
    plt.plot(t_cont, x_cont, label="Continuous Sinusoid")
    plt.title(f"Continuous Sinusoidal Signal (f = {signal_freq} Hz)")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Plot: Sampled version
    plt.figure(figsize=(10, 4))
    plt.stem(t_samples, x_samples, basefmt=" ", label="Sampled Points")
    plt.title(f"Sampled Sinusoid at {sampling_freq} Hz (Signal f = {signal_freq} Hz)")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return t_samples, x_samples

# Example usage
t, x = sample_sinusoid(signal_freq=50, sampling_freq=500)
```
