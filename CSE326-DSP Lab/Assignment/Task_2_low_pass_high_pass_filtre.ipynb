Problem statement: Design a Low-Pass Filter and a High-Pass Filter using suitable techniques (e.g., FIR filter design with windowing method) and analyze their frequency responses.

Required Library

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import firwin, freqz
```

Implementation

```python
def design_filters(fs=1000, cutoff_lp=200, cutoff_hp=300, numtaps=101):
    """
    Design low-pass and high-pass FIR filters.

    Parameters:
        fs: Sampling frequency (Hz)
        cutoff_lp: Cutoff for low-pass filter (Hz)
        cutoff_hp: Cutoff for high-pass filter (Hz)
        numtaps: Number of filter taps (filter length)

    Returns:
        lp_filter: Low-pass filter coefficients
        hp_filter: High-pass filter coefficients
    """
    # Design filters
    lp_filter = firwin(numtaps, cutoff=cutoff_lp, fs=fs, window='hamming')
    hp_filter = firwin(numtaps, cutoff=cutoff_hp, fs=fs, pass_zero=False, window='hamming')

    # Frequency response
    w_lp, h_lp = freqz(lp_filter, worN=8000, fs=fs)
    w_hp, h_hp = freqz(hp_filter, worN=8000, fs=fs)

    # Plot low-pass filter
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(w_lp, 20 * np.log10(np.abs(h_lp)))
    plt.title("Low-Pass Filter Frequency Response")
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Magnitude (dB)")
    plt.grid()

    # Plot high-pass filter
    plt.subplot(1, 2, 2)
    plt.plot(w_hp, 20 * np.log10(np.abs(h_hp)))
    plt.title("High-Pass Filter Frequency Response")
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Magnitude (dB)")
    plt.grid()

    plt.tight_layout()
    plt.show()

    return lp_filter, hp_filter

# Example usage
lp, hp = design_filters(fs=1000, cutoff_lp=200, cutoff_hp=300, numtaps=101)
```
