# Perform convolution sum 

```python
x = [2,3,1,5]
h = [1,2,3]
l = len(x)+len(h) - 1
y = [0]*l
# for i in range(len(x)):
#     for j in range(len(h)):
#         y[i+j] = y[i+j] + x[i]*h[j]
for n in range(len(x) + len(h) - 1):
    for k in range(len(x)):
      if 0 <= n - k < len(h):
        y[n] += x[k] * h[n - k]
print(y)        
print(np.convolve(x,h))        
```

# Cross Correlation

```python
#Cross-correlation
x = [2,3,1,5]
h = [1,2,3]
l = len(x)+len(h) - 1
y = [0]*l
hh = np.flip(h)
# for i in range(len(x)):
#     for j in range(len(h)):
#         y[i+j] = y[i+j] + x[i]*hh[j]
for n in range(len(x) + len(h) - 1):
    for k in range(len(x)):
      if 0 <= n - k < len(h):
        y[n] += x[k] * hh[n - k]
print(y)        
print(np.correlate(x,h, 'full')) 
```

# auto correlation

```python
#Auto correlation
x = [2,3,1,5]
h = [1,2,3]
l = len(x)+len(x) - 1
y = [0]*l
xh = np.flip(x)
# for i in range(len(x)):
#     for j in range(len(x)):
#         y[i+j] = y[i+j] + x[i]*xh[j]
for n in range(len(x) + len(x) - 1):
    for k in range(len(x)):
      if 0 <= n - k < len(x):
        y[n] += x[k] * xh[n - k]
print(y)        
print(np.correlate(x,x, 'full')) 
```

# Plot composite signal 

```python
# Define sampling parameters
Fs = 100  # Sampling frequency in Hz
T = 1 / Fs  # Sampling period
L = 500  # Number of samples
t = np.arange(0, L) * T  # Time vector

# Define signal parameters
A1, f1 = 2, 2   # Amplitude and frequency of the first sine wave
A2, f2 = 1, 6   # Amplitude and frequency of the second sine wave
A3, f3 = 0.5, 12  # Amplitude and frequency of the third sine wave

# Generate sine waves
y1 = A1 * np.sin(2 * np.pi * f1 * t)
y2 = A2 * np.sin(2 * np.pi * f2 * t)
y3 = A3 * np.sin(2 * np.pi * f3 * t)

# Composite signal (sum of the three sine waves)
y_composite = y1 + y2 + y3

# Plot composite signal
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t, y_composite, label="Composite Signal", color='b')
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Composite Signal (Sum of Three Sine Waves)")
plt.legend()
plt.grid(True)
```
